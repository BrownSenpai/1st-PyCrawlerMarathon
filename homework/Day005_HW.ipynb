```python
# 1. é€™å€‹ API ä¸€æ¬¡æœƒå›å‚³å¹¾ç­†è³‡æ–™ï¼Ÿæ¯ä¸€ç­†è³‡æ–™åŒ…å«å“ªäº›æ¬„ä½ï¼Ÿ
import requests
r=requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
r.text
```

```python
import json
data=json.loads(r.text)
data
```

```python
print(len(data))
for key in data[0].keys():
    print(key)
    
```

```python
# 2. å–å‡ºæ¯ä¸€ç­†è³‡æ–™çš„ã€Œæ¨™é¡Œã€ã€ã€Œè²¼æ–‡æ™‚é–“ã€ã€ã€Œç•™è¨€äººæ•¸ã€ã€ã€ŒæŒ‰è®šäººæ•¸ã€
for f in data:
    print('æ¨™é¡Œ:',f['title'],'è²¼æ–‡æ™‚é–“:',f['createdAt'],'ç•™è¨€äººæ•¸:',f['commentCount'],'æŒ‰è®šäººæ•¸:',f['likeCount'])
```

```python
# 3. è¨ˆç®—ç†±é–€/éç†±é–€æ–‡ç« çš„ã€Œå¹³å‡ç•™è¨€äººæ•¸ã€èˆ‡ã€Œå¹³å‡æŒ‰è®šäººæ•¸ã€
comcount_pop=0
likcount_pop=0
totalcount=0
for pc in data:
    comcount_pop+=pc['commentCount']
    likcount_pop+=pc['likeCount']
    totalcount+=1
print('ç†±é–€æ–‡ç« ç¸½ç•™è¨€äººæ•¸',comcount_pop)
print('ç†±é–€æ–‡ç« ç¸½æŒ‰è®šäººæ•¸',likcount_pop)
print('ç¯‡æ•¸:',totalcount)
print('ç†±é–€æ–‡ç« å¹³å‡ç•™è¨€äººæ•¸:',comcount_pop/totalcount)
print('ç†±é–€æ–‡ç« å¹³å‡æŒ‰è®šäººæ•¸:',likcount_pop/totalcount)
```
