## 作業目標

* 請利用 API: https://www.dcard.tw/_api/forums/pet/posts?popular=true 回答下列問題：

1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？
2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」
3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」

```python
# 1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？
import requests
r=requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
r.text
```

```python
import json
data=json.loads(r.text)
data
```

```python
print(len(data))
for key in data[0].keys():
    print(key)
    
```

```python
# 2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」
for f in data:
    print('標題:',f['title'],'貼文時間:',f['createdAt'],'留言人數:',f['commentCount'],'按讚人數:',f['likeCount'])
```

```python
# 3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」
comcount_pop=0
likcount_pop=0
totalcount=0
for pc in data:
    comcount_pop+=pc['commentCount']
    likcount_pop+=pc['likeCount']
    totalcount+=1
print('熱門文章總留言人數',comcount_pop)
print('熱門文章總按讚人數',likcount_pop)
print('篇數:',totalcount)
print('熱門文章平均留言人數:',comcount_pop/totalcount)
print('熱門文章平均按讚人數:',likcount_pop/totalcount)
```

```python
import requests
un=requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=false')
import json
dataun=json.loads(un.text)
dataun
```

```python
comcount_unpop=0
likcount_unpop=0
totalcount_un=0
for pc in dataun:
    comcount_unpop+=pc['commentCount']
    likcount_unpop+=pc['likeCount']
    totalcount_un+=1
print('非熱門文章總留言人數',comcount_unpop)
print('非熱門文章總按讚人數',likcount_unpop)
print('篇數:',totalcount_un)
print('非熱門文章平均留言人數:',comcount_unpop/totalcount_un)
print('非熱門文章平均按讚人數:',likcount_unpop/totalcount_un)
```
