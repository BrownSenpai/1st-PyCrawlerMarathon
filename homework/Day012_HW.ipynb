### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html5lib")
for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find(class_="date").text,d.find(class_="tag").text,d.find_all('a')[-1].text)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
classify_={}

for d in soup.find(class_="part_list_2").find_all('h3'):
    class_tag=d.select('.tag')[0].text
    title_=d.find(class_="date").text +' '+d.find_all('a')[-1].text
    classify_.setdefault(class_tag,[])
    classify_[class_tag].append({title_})
print(classify_)
```

### ③ 哪一個類別的文章最多

```python
classtag=[]
class_statics={}
for d in soup.find(class_="part_list_2").find_all('h3'):
    class_tag=d.find(class_='tag').text
    print(class_tag)
    classtag.append(class_tag)
for i in classtag:
    if i not in class_statics:
            class_statics[i]=1
    else:
            class_statics[i]+=1
print(class_statics)
```

```python
print(sorted(class_statics.items(), key=lambda class_statics:class_statics[1],reverse=True))

highest = max(class_statics.values())
print([k for k, v in class_statics.items() if v == highest])
```
