### 1. Dcard 網址： https://www.dcard.tw/f

### 2. 知乎： https://www.zhihu.com/explore

### 3. 觀察一下知乎回來的資料好像有點怪怪的，該怎麼解決？

```python
import requests
url = 'https://www.zhihu.com/explore'
#加上headers
headers={'user-agent': 'my-app/0.0.1'}
r_2 = requests.get(url,headers=headers)

r_2.encoding = 'utf-8'
print(r_2.text[0:600])
```

```python
soup_r_2=BeautifulSoup(r_2.text, "html5lib")
print(soup_r_2)
```
